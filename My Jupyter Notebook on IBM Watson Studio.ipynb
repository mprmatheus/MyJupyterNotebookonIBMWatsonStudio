### The following code is supposed to calculate the volume of a cylinder with 5 cm of diameter and 10 cm of height.

```python
pi = 3.14
diameter = 5
height = 10
cylinder_volume = pi*((pow(diameter,2))/4)*height
print("The volume of the given cylinder is " + str(cylinder_volume) + " cm³")
```
